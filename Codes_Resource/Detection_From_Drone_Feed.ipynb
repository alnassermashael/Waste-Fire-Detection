```python
import cv2
from ultralytics import YOLO
```

```python
# Load the model
model = YOLO("/content/drive/MyDrive/training_results/train55/weights/best.pt", task = "detect")
```

```python
# Create Stream Predection object
res = model.predict(source = "rtmp://127.0.0.1/live/k", stream = True, show = True)
```

```python
# Show the the detection frames.
for r in res:
    cv2.imshow("WIn", r.plot())
    cv2.waitKey(1)
```
