```python
import os
import pandas as pd
```

```python
for label in os.listdir():
    if(not label.endswith(".txt")):
        pass
    else:
        # Open the label txt file
        f = open(label,"r")

        # Save the cords of the label in a list.
        cords = list(map(lambda cord: cord,f.readline().split(" ")))
        f.close()

        # Change the class number(The first number in the cords) to 1.
        cords[0] = 1

        # Open the same label file to overwrite on it.
        f = open(label,"w")

        for cord in cords:
            f.write(str(cord))
            f.write(" ")
        f.close()
```
