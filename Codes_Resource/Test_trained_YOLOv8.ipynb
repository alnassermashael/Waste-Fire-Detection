```python
from ultralytics import YOLO
from matplotlib import pyplot as plt
import os
import cv2
```

```python
# Load the model
model = YOLO("/content/drive/MyDrive/training_results/Fire_Smoke_Detection.v6/weights/best.pt")
```

```python
# A list to contain all the testing images names.
imgs = []

# Collect all the images names and add them to the imgs list.
for elm in os.listdir("/content/drive/MyDrive/Data/test/images"):
    if(elm.endswith("jpg")):
        imgs.append(f"/content/drive/MyDrive/Data/test/images/{elm}")
```

```python
# Run inference on the all the saved images.
results = model(imgs)  # return a list of Results objects
```

```python
plt.imshow(cv2.cvtColor(results[5].plot(), cv2.COLOR_BGR2RGB))
```
